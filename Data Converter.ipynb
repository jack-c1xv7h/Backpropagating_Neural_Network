```python
import csv
import numpy as np

with open("output.csv", mode="w", newline="") as csv_file:
    writer = csv.writer(csv_file)

    for i in range(5):
         with open("save" + str(i+1) + ".txt", mode='r') as file:
            txt = file.readlines()
            for linecount, line in enumerate(txt):
                textlength = len(txt)
                if linecount != textlength - 1:
                    values = np.array(line.split())
                    values = np.array(values, dtype=np.float32)
                    
                    #for v in range(textlength-1):
                    #    line = file.readline(v)
                    #    print(line)
                    #    print(linearray, i, v+1)

                print(writer, textlength, values)
                
```
